```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize = (18, 6))
sns.set(style = 'darkgrid')
sns.countplot(data = music_data, x = 'year', color = 'blue')
plt.title('Distribution of Tracks Across Decades')
plt.show()
```

```python
fig = px.line(data_by_year, x='year', y= ['instrumentalness','acousticness','loudness'], title='Trend of various sound features over decades')
fig.show()
```

```python
fig = px.line(data_by_year, x='year', y='loudness', title='Trend of loudness over decades')
fig.show()
```

```python
sel = gen_data[['genres', 'valence', 'energy', 'danceability', 'acousticness']] # Select the necessary columns from gen_data
sel = sel.groupby('genres')[['valence', 'energy', 'danceability', 'acousticness']].mean().round(2)\
          .sort_values(by = ['valence', 'energy', 'danceability', 'acousticness'], ascending = False)


sel = sel.head(10)

fig = px.bar(sel, x=sel.index, y=['valence', 'energy', 'danceability', 'acousticness'], barmode='group', title='Trend of various sound features over top 10 genres')
#                                 ^ Changed sel.reset_index to sel.index
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Load your dataset (modify the path if needed)

# Check the column names of art_data
print(art_data.columns)

# Extract artist names and combine them into a single string
# Replace 'artist' with the actual column name from the printed output
# For example if the column is names 'artists' use art_data['artists'] instead
comment_words = " ".join(art_data["artists"].astype(str)) # If the column name is 'artists'

# Define stopwords
stopwords = set(STOPWORDS)

# Generate WordCloud
wordcloud = WordCloud(width=800, height=800, background_color='white',
                      stopwords=stopwords, min_word_length=3, max_words=40,
                      min_font_size=10).generate(comment_words)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Count number of songs per artist
# Replace 'artists' with the correct column name if needed
top10_most_song_produced_artists = art_data["artists"].value_counts().reset_index()

# Rename columns for clarity
top10_most_song_produced_artists.columns = ["artists", "count"]

# Display the top 10 artists
top10_most_song_produced_artists = top10_most_song_produced_artists.head(10)

# Print the result
print(top10_most_song_produced_artists[['count', 'artists']].sort_values('count', ascending=False))
```

```python

# Sort by popularity in descending order
top10_popular_artists = art_data[['artists', 'popularity']].sort_values('popularity', ascending=False)

# Get the top 10 artists
top10_popular_artists = top10_popular_artists.head(10)

# Display the top 10 artists with the highest popularity scores
print(top10_popular_artists)
```
